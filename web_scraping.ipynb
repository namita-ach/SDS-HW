```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = "https://www.amazon.in/NutriPro-Bullet-Juicer-Grinder-Blades/product-reviews/B09J2SCVQT"
```

```python
page = requests.get(link)
print(page)
```

```python
soup = bs(page.content, 'html.parser')
print(soup.prettify())
```

```python
name = soup.find_all('span', class_='a-profile-name')
name
```

```python
cust_name =[]
for i in range(0, len(name)):
    cust_name.append(name[i].get_text())
cust_name
```

```python
cust_name.pop(11)
cust_name.pop(5)
cust_name.pop(11)
cust_name.pop(0)
```

```python
len(cust_name)
```

```python
cust_name
```

```python
title = soup.find_all('a', class_= 'review-title-content')
title
```

```python
review_title=[]
for i in range(0, len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
len(review_title)
```

```python
review_title[:]= [titles.lstrip('\n') for titles in review_title]
review_title[:]= [titles.rstrip('\n') for titles in review_title]
review_title[:]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
df['Customer name'] = cust_name
df['Review title'] = review_title
df
```

```python
# determining the name of the file
file_name = 'review.xlsx'
  
# saving the excel
if df.to_excel(file_name):
    print('DataFrame is written to Excel File successfully')
```
