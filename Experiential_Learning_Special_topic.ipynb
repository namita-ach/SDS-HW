```r
#convert it to pure numbers
listings$price <- as.numeric(gsub("\\$|,", "", listings$price))
print(class(listings$price))
```

```r
#we check fro the correlation between price (which I'm taking as dependent variable) and each of the numeric parameters

price_correlations <- c()

for (col_name in names(listings)) {
  if (column_data_types[col_name] == "numeric" && col_name != "price") {
    #check and handle NA values for both columns
    numeric_values_price <- as.numeric(listings$price)
    numeric_values_col <- as.numeric(listings[[col_name]])

    if (!any(is.na(numeric_values_price)) && !any(is.na(numeric_values_col))) {
      #calculate correlation with "price"
      correlation <- cor(numeric_values_price, numeric_values_col, use = "complete.obs")

      #store correlation coefficient
      price_correlations <- c(price_correlations, correlation)

      cat("Correlation between price and", col_name, ":", correlation, "\n")
    }
  }
}

cat("Correlation coefficients:", price_correlations, "\n")
```

```r
#The very low p value indicates (along with the scatterplot) that the relation is far from linear
#We can use a generalized additive model (GAM) to help fit non-linear data

install.packages("mgcv")
```

```r
#load the package
library(mgcv)
```

```r
#make a model
model_latitude_gam <- gam(price ~ s(latitude), data = listings)
summary(model_latitude_gam)

'''
The low p-value for the smooth term indicates that latitude has a significant non-linear relationship with price.
The low adjusted R-squared suggests that the model might not capture a large portion of the variability in price.
The number of degrees of freedom is 8.535, suggesting a relatively complex relationship.
'''
```

```r
plot(model_latitude_gam, se = TRUE, col = "purple")

#positive and statistically significant relationship between latitude and price
#price increases as latitude increases- it is not by chance
#the relationship is non-linear
#we are 95% confident that the true gradient of the model's latitude is within the shaded area
#there is less data available at higher latitude- which is why the model is less confident
```

```r
model_numRev_gam <- gam(price ~ s(number_of_reviews), data = listings)
summary(model_numRev_gam)

'''
The low p-value for the smooth term suggests that number_of_reviews has a statistically significant non-linear relationship with price.
The adjusted R-squared is very low, indicating that the model explains only a tiny proportion of the variability in price
The number of degrees of freedom is 5.067.
'''
```

```r
plot(model_numRev_gam, se = TRUE, col = "blue")

#number of reviews has a positive and statistically significant effect on price
#model predicts that price will increase as the number of reviews increases
#rate of increase in price decreases as the number of reviews increases
#less data available at higher numbers of reviews- model is not confident
```

```r
model_accomm_gam <- gam(price ~ s(accommodates), data = listings)
summary(model_accomm_gam)

'''
The low p-value for the smooth term suggests that accommodates has a statistically significant non-linear relationship with price.

The adjusted R-squared, while still relatively low, is higher compared to the previous models.
It suggests that the model explains a larger proportion of the variability in price.

The number of degrees of freedom is 8.973.
'''
```

```r
plot(model_accomm_gam, se = TRUE, col = "green")

#number of accommodations has a positive and statistically significant effect on price
#the relationship is non-linear
#price will increase as the number of accommodations increases
#rate of increase in price decreases as the number of accommodations increases

#adding more accommodations does not lead to a significant increase in price
```
