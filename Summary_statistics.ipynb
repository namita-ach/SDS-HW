

> Using median to replace null values



```python
def fill_na_with_median(df, name_of_col):
  median = np.nanmedian(df[name_of_col])
  df[name_of_col].replace(np.nan, median, inplace=True)
  return df
```

```python
fill_na_with_median(df_for_median, 'sealevelpressure')
fill_na_with_median(df_for_median, 'tempmax')
fill_na_with_median(df_for_median, 'tempmin')
fill_na_with_median(df_for_median, 'temp')
fill_na_with_median(df_for_median, 'feelslikemax')
fill_na_with_median(df_for_median, 'feelslikemin')
fill_na_with_median(df_for_median, 'feelslike')
fill_na_with_median(df_for_median, 'dew')
fill_na_with_median(df_for_median, 'windspeed')
fill_na_with_median(df_for_median, 'humidity')
```

```python
no_of_clusters = 3
unique = list(set(df_for_mean['tempmin']))
smp = sample(unique, no_of_clusters)
print("clusters:", smp, "selected out of:", unique, '\n')

# Create index sets for each cluster
index_set = [list(df_for_mean.index[df_for_mean['tempmin'] == i]) for i in smp]

# Flatten the index sets to obtain the selected indices
selected_indices = [j for i in index_set for j in i]

# Select rows from the original DataFrame using the selected indices
selected_data = df_for_mean.loc[selected_indices]

print(selected_data)
```

```python
no_of_elements = 2
unique = list(set(df_for_mean['tempmax']))
print('stratas:', unique, '\n')

# Sample indices based on stratified conditions
index_set = [
    sample(
        list(df_for_mean.index[(df_for_mean['tempmin'] == i) & (df_for_mean['feelslikemax'] == i)]),
        min(no_of_elements, len(df_for_mean[(df_for_mean['tempmin'] == i) & (df_for_mean['feelslikemax'] == i)]))
    )
    for i in unique
]

# Flatten the index sets to obtain the selected indices
selected_indices = [j for i in index_set for j in i]

# Select rows from the original DataFrame using the selected indices
selected_data = df_for_mean.loc[selected_indices]

print(selected_data)
```

```python
def npp(data):
    data = sorted(data)
    p = [(data.index(i)-0.5)/len(data) for i in data]
    z = zscore(p)

    plt.scatter(data, z)
    plt.ylabel('Z values')
    plt.xlabel(' values')
    plt.show()
    plt.plot(data,  z,'ro',data,  data)
    plt.show()
```
