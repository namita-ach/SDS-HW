Sampling

```python
from random import sample
from statistics import mean
```

```python
#simple random sampling
no_of_elements = 10
random_index = sample(range(df.shape[0]), no_of_elements)
print(random_index)
print(df.iloc[random_index])
```

```python
#systematic sampling
kth = 10
index = [i for i in range(df.shape[0]) if i%kth==0]
print(df.iloc[index])
```

```python
#stratified sampling
#stratas formed based on total profit
no_of_elements = 2      #number of elements in each strata
unique = list(set(df['total_profit']))
print('stratas :', unique, '\n')
index_set = [sample(list(df.index[df['total_profit'] == i]), min(no_of_elements, len(df[df['total_profit'] == i]))) for i in unique]
index = [j for i in index_set for j in i]
print(df.iloc[index])
```

```python
#cluster sampling
#clusters formed based on number of total units sold
no_of_clusters = 3
unique = list(set(df['total_units']))
smp = sample(unique, no_of_clusters)
print("clusters :", smp, "selected out of :", unique, '\n')
index_set = [list(df.index[df['total_units']==i]) for i in smp]
index = [j for i in index_set for j in i]
print(df.iloc[index])
```
