```python
#central limit theorem with 100 elements each in 1000 samples for the parameter 'tempmax'
no_of_elements = 100
num_samples = 1000

#Collect means from multiple random samples
means = []
for _ in range(num_samples):
    random_index = sample(range(df_for_mean.shape[0]), no_of_elements)
    sample_mean = mean(df_for_mean.iloc[random_index]['tempmax'])
    means.append(sample_mean)

plt.hist(means, bins=30)
plt.xlabel('Mean Values')
plt.ylabel('Frequency')
plt.title('Histogram of Means')
plt.show()
```

```python
#normal probability plot
z_scores = zscore(means)
plt.scatter(means, z_scores)
plt.xlabel('Mean Values')
plt.ylabel('Z scores')
plt.title('Scatter Plot of Means vs Z scores')
plt.show()

plt.plot(means, z_scores, 'ro', label='Z scores')
plt.plot(means, means, label='Mean Values')
plt.xlabel('Mean Values')
plt.ylabel('Z scores')
plt.title('Line Plot of Means and Z scores')
plt.legend()
plt.show()
```

```python
def npp(data):
    data = sorted(data)
    p = [(data.index(i)-0.5)/len(data) for i in data]
    z = zscore(p)

    plt.scatter(data, z)
    plt.ylabel('Z values')
    plt.xlabel(' values')
    plt.show()
    plt.plot(data,  z,'ro',data,  data)
    plt.show()
```

```python
#qq plot
no_of_elements = 100
num_samples = 1000

qq_sample = []
for _ in range(num_samples):
    random_index = sample(range(df_for_mean.shape[0]), no_of_elements)
    qq_sample.append(mean(df_for_mean.iloc[random_index]['tempmax']))

def qq_plot(data):
    probplot(data, dist="norm", plot=plt)
    plt.title("Q-Q Plot")
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Sample Quantiles")
    plt.show()

qq_plot(qq_sample)
```

```python
data = df_for_mean['tempmax']
mu, std = norm.fit(data)

# Plot the histogram
plt.hist(data, bins=25, density=True, alpha=0.6, color='b')

# Plot the best fit line
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

# Add labels
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

## Unit 3

Confidence intervals, hypothesis testing, and effects of various parameters on margin of error

```python
from scipy.stats import t, norm
import math
```

```python
qq_sample = np.random.normal(size=100)

#Plotting the sample data
plt.hist(qq_sample, bins=30, density=True, alpha=0.6, color='black', label='Sample Data')

#Plotting the t-distribution curves
for df in [1, 2, 5]:
    x_values = np.linspace(qq_sample.min(), qq_sample.max(), 100)
    plt.plot(x_values, t.pdf(x_values, df), '--', label=f't-distribution (df={df})')

#Plotting the standard normal curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
plt.plot(x, norm.pdf(x), 'k', linewidth=2, label='Std Normal Curve')

plt.legend(loc='upper right')
plt.title('Histogram and Standard Normal Curve')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()

#print probabilities using t and normal distributions
print("1-cdf gives:", 1 - t.cdf(1.59, 2))
print('Same as:', t.sf(1.59, 2))
print(1 - norm.cdf(2), norm.sf(2))
```

```python
def conf_int(t_score, df):
    plt.figure(figsize=(8, 4))
    x = np.linspace(-6, 6, 1000)

    #Calculate the confidence level
    area = t.cdf(t_score, df) - t.cdf(-t_score, df)
    print('Confidence Level:', area * 100)

    #Plot the t-distribution curve
    plt.plot(x, t.pdf(x, df), color='red', label=f't-distribution (df={df})')

    #Fill the area between -t_score and +t_score
    section = np.arange(-t_score, t_score, 1/20.)
    plt.fill_between(section, t.pdf(section, df), alpha=0.3)

    plt.xticks(np.arange(-6, 7, 0.5), rotation=45)
    plt.legend(loc='upper right')
    plt.title('T-Distribution and Confidence Interval')
    plt.xlabel('Values')
    plt.ylabel('Density')
    plt.show()

conf_int(1.96, 2)
```

```python
#margin of error
qq_sample = np.random.normal(size=100)
```

```python
#Effect of Standard deviation on Margin of error
#MOE = t_alpha * sigma/sqrt(n)
confidence_level = 0.95
n = len(qq_sample)
moe = []
x = np.arange(0, 21, 1)  #vary SD from 0 to 20

for sd in x:
    moe.append(t.ppf((1 + confidence_level) / 2, n - 1) * sd / math.sqrt(n))

plt.plot(x, moe)
plt.title(f'n = {n}, CL = {confidence_level * 100}% and Varying the Standard deviation')
plt.ylabel('Margin of error')
plt.xlabel('Standard deviation')
plt.show()
```

```python
#effect of sample size on moe

sigma = 10
z_score = 1.96  #95% confidence level
x = range(20, 801)  #varying sample size from 20 to 800

moe = []
for n in x:
    moe.append(1.96 * sigma / math.sqrt(n))

plt.plot(x, moe)
plt.title('sigma = 10, CL = 95% and varying sample size')
plt.ylabel('Margin of error')
plt.xlabel('Sample size')
plt.show()
```

```python
#effect of confidence level on moe

sigma = 10
z_score = 1.96  #95% confidence level
n=64

def prob(z1,z2):
       return (norm.cdf(z2) - norm.cdf(z1))


z_alphaby2 = np.linspace(norm.ppf(0.10), norm.ppf(0.0005), 100)
z_alphaby2 = z_alphaby2 * -1

#calculating margin of error
moe = z_alphaby2 * sigma/math.sqrt(n)

#getting the CL
x = []
for i in range(0, len(z_alphaby2)):
    x.append(prob(-1 * z_alphaby2[i], z_alphaby2[i]))

#getting the CL in %
x = np.array(x) * 100

plt.plot(x, moe)
plt.ylabel('Margin of error')
plt.ylim(0.5,5)
plt.xlim(80,100)
plt.title('sigma = 10, n = 64 and Varying the CL')
plt.xlabel('confidence level')
plt.show()
```

```python
#hypothesis testing

'''
Hypothesis:
The humidity levels for the day have no impact on the maximum temperature (tempmax)
'''

#H(0): Humidity has a significant impact on tempmax
#H(1): Humidity has no impact on tempmax

alpha = 0.05

#Calculate Pearson correlation coefficient and p-value
corr_coeff, p_value = stats.pearsonr(df_for_mean['humidity'], df_for_mean['tempmax'])

print(f"Pearson Correlation Coefficient: {corr_coeff}")
print(f"P-value: {p_value}")

#Visualize the relationship with a scatter plot
plt.scatter(df_for_mean['humidity'], df_for_mean['tempmax'], s=10)
plt.title('Scatter Plot: Humidity vs tempmax')
plt.xlabel('Humidity')
plt.ylabel('tempmax')
plt.show()

#Check the p-value against the significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant correlation between Humidity and tempmax.")
else:
    print("Fail to reject the null hypothesis. There is no significant correlation between Humidity and tempmax.")
```

## Unit 4
Simple linear regression and residual plots

```python
#linear regression
y = df_for_mean['tempmax'] #humidity is our independednt variable
x = df_for_mean['humidity'] #temp is dependent variable
n = np.size(x)

#mean of x and y
mean_x = np.sum(x)/n
mean_y = np.sum(y)/n

#summation values taken
summation_x_y = np.sum(y*x) - n*mean_y*mean_x
summation_x_x = np.sum(x*x) - n*mean_x*mean_x
b_1 = summation_x_y /summation_x_x
b_0 = mean_y - b_1*mean_x

print("Estimated coefficients:\nb_0 =",b_0," \nb_1 =" ,b_1)
plt.scatter(x, y, s=5, color='pink', label='Data Points')
y_predicted = b_0 + b_1*x
plt.plot(x, y_predicted, color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
